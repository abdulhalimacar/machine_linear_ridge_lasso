```python
from sklearn.preprocessing import OneHotEncoder,OrdinalEncoder
```

```python
X = df.drop(columns="price")
y = df["price"]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=101)
```

```python
cat_onehot = ['make_model', 'body_type', 'Type', 'Fuel', 'Paint_Type','Upholstery_type', 'Gearing_Type', 'Drive_chain']
cat_ordinal = ['Comfort_Convenience_Package', 'Entertainment_Media_Package', 'Safety_Security_Package']
```

```python
cat_for_comfort = ['Standard', 'Premium', 'Premium Plus']
cat_for_ent = ['Standard', 'Plus']
cat_for_safety = ['Safety Standard Package', 'Safety Premium Package', 'Safety Premium Plus Package']
```

```python
from sklearn.compose import make_column_transformer
```

```python
enc_onehot = OneHotEncoder(handle_unknown="ignore", sparse=False)
enc_ordinal = OrdinalEncoder(categories= [cat_for_comfort, cat_for_ent, cat_for_safety])
```

```python
column_trans = make_column_transformer((enc_onehot, cat_onehot), 
                                       (enc_ordinal, cat_ordinal),
                                       remainder='passthrough')
```

```python
#eÄŸer standard scaling kullanÄ±rsak bu bize datanÄ±n standart sapmasÄ±nÄ± 1'e ve mean'ini sÄ±fÄ±ra eÅŸitler
#hangi scaling kullanÄ±p kullanmayacaÄŸÄ±mÄ±za ÅŸÃ¶yle karar veriririz: her iki skaling sÄ±rasÄ±yla denenir hangisi iyi sonuÃ§ verirse
#onunla devam edilir
```

# LINEAR REGRESSION

```python
operations = [("Encoding",column_trans),("Scaling",MinMaxScaler()),("Linear",LinearRegression())]
```

```python
model = Pipeline(steps=operations)
```

```python
model.fit(X_train,y_train)
```

```python
y_test_pred = model.predict(X_test)
y_train_pred = model.predict(X_train)
```

```python
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
```

```python
r2_test = r2_score(y_test,y_test_pred)
r2_train = r2_score(y_train,y_train_pred)
rmse_test = np.sqrt(mean_squared_error(y_test,y_test_pred))
rmse_train = np.sqrt(mean_squared_error(y_train,y_train_pred))
```

```python
my_dict = {"r2_test":r2_test,"r2_train":r2_train,"rmse_test":rmse_test,"rmse_train":rmse_train}
my_dict.values()
```

```python
pd.DataFrame(data=my_dict.values(),index=my_dict,columns=["results_linear"])
```

```python
#encoding iÅŸlemi ile bizim birsÃ¼rÃ¼ yeni sÃ¼tunumuz oldu ve sÃ¼tun artmasÄ± kesin kesin olarak r2_skorunu iyileÅŸtirir. Bunun kont-
#rolÃ¼nÃ¼ Adjusted_R2 ile yaparÄ±z
```

```python
def adj_r2(y_test, y_pred, X):
    r2 = r2_score(y_test, y_pred)
    n = X.shape[0]   # number of observations
    p = X.shape[1]   # number of independent variables 
    adj_r2 = 1 - (1-r2)*(n-1)/(n-p-1)
    return adj_r2
```

```python
adj_r2(y_test,y_test_pred,X)
```

```python
#adjusted_r2 skoru ile r2_skoru gÃ¶rÃ¼ldÃ¼ÄŸÃ¼ Ã¼zere birbirine oldukÃ§a yakÄ±n Ã§Ä±kmÄ±ÅŸtÄ±r.
```
